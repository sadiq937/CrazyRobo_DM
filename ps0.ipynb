```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set up Seaborn for better visualizations
sns.set(style="darkgrid")

# Read CSV file (sales_pipeline.csv)
sales_df = pd.read_csv('sales_pipeline.csv')

# Display the first few rows of the dataset
print("Sales Pipeline Data (CSV):")
print(sales_df.head())

# Function to display basic descriptive statistics
def display_basic_stats(df, name):
    print(f"\nBasic Descriptive Statistics for {name}:")
    print(df.describe(include='all')) 
    print("\nMissing Values:")
    print(df.isnull().sum())

display_basic_stats(sales_df, "Sales Pipeline Data")

# Plot a histogram for the 'close_value' column
plt.figure(figsize=(12, 8))
sns.histplot(sales_df['close_value'], kde=True, color='green', bins=10)
plt.title("Distribution of Close Value (USD)")
plt.xlabel("Close Value (USD)")
plt.ylabel("Frequency")
plt.show()
```
